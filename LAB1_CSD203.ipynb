```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.last = None

    def addToHead(self, x):
        new_node = Node(x)
        if not self.last:
            new_node.next = new_node  
            self.last = new_node
        else:
            new_node.next = self.last.next
            self.last.next = new_node

    def addToTail(self, x):
        new_node = Node(x)
        if not self.last:
            new_node.next = new_node  
            self.last = new_node
        else:
            new_node.next = self.last.next
            self.last.next = new_node
            self.last = new_node

    def addAfter(self, p, x):
        if not p:
            print("Invalid node reference")
            return

        new_node = Node(x)
        new_node.next = p.next
        p.next = new_node

    def traverse(self):
        if not self.last:
            print("List is empty")
            return

        temp = self.last.next
        while True:
            print(temp.data, end=' ')
            temp = temp.next
            if temp == self.last.next:
                break
        print()

    def deleteFromHead(self):
        if not self.last:
            print("List is empty")
            return None

        deleted_data = self.last.next.data
        if self.last.next == self.last:
            self.last = None
        else:
            self.last.next = self.last.next.next

        return deleted_data

    def deleteFromTail(self):
        if not self.last:
            print("List is empty")
            return None

        deleted_data = None
        if self.last.next == self.last:
            deleted_data = self.last.data
            self.last = None
        else:
            temp = self.last.next
            while temp.next != self.last:
                temp = temp.next
            deleted_data = self.last.data
            temp.next = self.last.next
            self.last = temp

        return deleted_data

    def deleteAfter(self, p):
        if not p or not p.next:
            print("Invalid node reference")
            return None

        deleted_data = p.next.data
        p.next = p.next.next

        return deleted_data

    def delete(self, x):
        if not self.last:
            print("List is empty")
            return None

        deleted_data = None
        if self.last.next.data == x:
            deleted_data = self.last.next.data
            if self.last.next == self.last:
                self.last = None
            else:
                self.last.next = self.last.next.next
        else:
            temp = self.last.next
            while temp.next != self.last.next and temp.next.data != x:
                temp = temp.next

            if temp.next != self.last.next:
                deleted_data = temp.next.data
                temp.next = temp.next.next

        return deleted_data

    def search(self, x):
        if not self.last:
            print("List is empty")
            return None

        temp = self.last.next
        while temp.next != self.last.next and temp.data != x:
            temp = temp.next

        if temp.data == x:
            return temp
        else:
            print("Node not found")
            return None

    def count(self):
        count = 0
        temp = self.last.next
        while True:
            count += 1
            temp = temp.next
            if temp == self.last.next:
                break
        return count
    
    def delAtIndex(self, i):
        if not self.last or i < 0:
            print("Invalid index or empty list")
            return None

        if i == 0:
            return self.deleteFromHead()

        temp = self.last.next
        prev = None
        count = 0

        while count < i:
            prev = temp
            temp = temp.next
            count += 1

        deleted_data = temp.data
        prev.next = temp.next

        return deleted_data

    def sort(self):
        if not self.last:
            print("List is empty")
            return

        current = self.last.next
        sorted_list = None

        while True:
            value = current.data
            sorted_list = self.insertSorted(sorted_list, value)
            current = current.next

            if current == self.last.next:
                break

        self.last = sorted_list

    def delNode(self, p):
        if not self.last or not p:
            print("Invalid node reference or empty list")
            return None

        if p == self.last:
            return self.deleteFromTail()

        temp = self.last.next
        prev = None

        while temp != p and temp.next != self.last.next:
            prev = temp
            temp = temp.next

        if temp == p:
            prev.next = temp.next
            return temp.data
        else:
            print("Node not found")
            return None

    def toArray(self):
        if not self.last:
            return []

        temp = self.last.next
        result = []

        while True:
            result.append(temp.data)
            temp = temp.next

            if temp == self.last.next:
                break

        return result

    def mergeSortedLists(list1, list2):
        merged_list = CircularLinkedList()

        current1 = list1.last.next
        current2 = list2.last.next

        while True:
            if not current1:
                merged_list.addToTail(current2.data)
                current2 = current2.next
            elif not current2:
                merged_list.addToTail(current1.data)
                current1 = current1.next
            elif current1.data < current2.data:
                merged_list.addToTail(current1.data)
                current1 = current1.next
            else:
                merged_list.addToTail(current2.data)
                current2 = current2.next

            if (current1 == list1.last.next) and (current2 == list2.last.next):
                break

        return merged_list

    def addBefore(self, p, x):
        if not self.last or not p:
            print("Invalid node reference or empty list")
            return

        new_node = Node(x)

        if p == self.last:
            self.addToTail(x)
            return

        temp = self.last.next
        prev = None

        while temp != p and temp.next != self.last.next:
            prev = temp
            temp = temp.next

        if temp == p:
            prev.next = new_node
            new_node.next = temp
        else:
            print("Node not found")

    def attachList(self, another_list):
        if not self.last:
            self.last = another_list.last
        elif another_list.last:
            temp = self.last
            self.last = another_list.last.next
            temp.next = another_list.last.next

    def max(self):
        if not self.last:
            print("List is empty")
            return None

        max_value = float('-inf')
        temp = self.last.next

        while True:
            max_value = max(max_value, temp.data)
            temp = temp.next

            if temp == self.last.next:
                break

        return max_value

    def min(self):
        if not self.last:
            print("List is empty")
            return None

        min_value = float('inf')
        temp = self.last.next

        while True:
            min_value = min(min_value, temp.data)
            temp = temp.next

            if temp == self.last.next:
                break

        return min_value

    def sum(self):
        if not self.last:
            print("List is empty")
            return 0

        total_sum = 0
        temp = self.last.next

        while True:
            total_sum += temp.data
            temp = temp.next

            if temp == self.last.next:
                break

        return total_sum

    def avg(self):
        count = self.count()
        if count == 0:
            return 0
        return self.sum() / count

    def sorted(self):
        if not self.last:
            print("List is empty")
            return True

        temp = self.last.next

        while temp.next != self.last.next:
            if temp.data > temp.next.data:
                return False
            temp = temp.next

        return True

    def insertSorted(self, sorted_list, x):
        new_node = Node(x)

        if not sorted_list or x < sorted_list.data:
            new_node.next = sorted_list
            return new_node

        temp = sorted_list

        while temp.next and temp.next.data < x:
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

        return sorted_list

    def reverse(self):
        if not self.last:
            print("List is empty")
            return

        prev = None
        current = self.last.next
        self.last = current

        while True:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

            if current == self.last.next:
                break

        self.last.next = prev

    def areEqual(list1, list2):
        current1 = list1.last.next
        current2 = list2.last.next

        while True:
            if not current1 and not current2:
                return True
            elif current1 and current2 and current1.data == current2.data:
                current1 = current1.next
                current2 = current2.next
            else:
                return False

            if (current1 == list1.last.next) and (current2 == list2.last.next):
                break

        return True


cll = CircularLinkedList()

cll.addToHead(1)
cll.addToHead(2)
cll.addToTail(3)
cll.addToTail(4)

cll.traverse()

print("Count:", cll.count())

node_to_search = cll.search(3)
if node_to_search:
    print("Found node with data:", node_to_search.data)

deleted_data = cll.delete(2)
print("Deleted node with data:", deleted_data)

cll.traverse()
print("Count:", cll.count())

cll1 = CircularLinkedList()
cll1.addToTail(3)
cll1.addToTail(5)
cll1.addToTail(7)

cll2 = CircularLinkedList()
cll2.addToTail(4)
cll2.addToTail(6)
cll2.addToTail(8)

print("List 1:")
cll1.traverse()

print("List 2:")
cll2.traverse()

merged_list = CircularLinkedList.mergeSortedLists(cll1, cll2)
print("Merged List:")
merged_list.traverse()

cll1.addBefore(cll1.search(5), 2)
print("List 1 after adding before node with value 5:")
cll1.traverse()

cll1.attachList(cll2)
print("List 1 after attaching List 2:")
cll1.traverse()

print("Max value in List 1:", cll1.max())
print("Min value in List 1:", cll1.min())
print("Sum of values in List 1:", cll1.sum())
print("Average of values in List 1:", cll1.avg())

print("Is List 1 sorted?", cll1.sorted())

cll1.sort()
print("List 1 after sorting:")
cll1.traverse()

cll1.insertSorted(9)
print("List 1 after inserting 9 into sorted list:")
cll1.traverse()

cll1.reverse()
print("List 1 after reversing:")
cll1.traverse()

print("Are List 1 and List 2 equal?", CircularLinkedList.areEqual(cll1, cll2))
```
