# Question 1

```python
class EmptyStackException(Exception):
    pass

class Node:
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next

class Stack:
    def __init__(self):
        self.top_node = None

    def isEmpty(self):
        return self.top_node is None

    def clear(self):
        self.top_node = None

    def push(self, x):
        new_node = Node(value=x, next=self.top_node)
        self.top_node = new_node

    def pop(self):
        if self.isEmpty():
            raise EmptyStackException("Stack is empty")
        value = self.top_node.value
        self.top_node = self.top_node.next
        return value

    def top(self):
        if self.isEmpty():
            raise EmptyStackException("Stack is empty")
        return self.top_node.value

    def traverse(self):
        current = self.top_node
        while current is not None:
            print(current.value, end=" ")
            current = current.next
        print()

def decimal_to_binary(decimal_number):
    stack = Stack()
    while decimal_number > 0:
        remainder = decimal_number % 2
        stack.push(remainder)
        decimal_number //= 2
    binary_result = ""
    while not stack.isEmpty():
        binary_result += str(stack.pop())
    return binary_result


stack_example = Stack()
print("Is stack empty?", stack_example.isEmpty())

stack_example.push(10)
stack_example.push(20)
stack_example.push(30)

print("Stack elements:")
stack_example.traverse()

print("Top element:", stack_example.top())

pop_value = stack_example.pop()
print("Popped value:", pop_value)

print("Is stack empty?", stack_example.isEmpty())

print("Converting 25 to binary:", decimal_to_binary(25))
stack_example.traverse()
```
